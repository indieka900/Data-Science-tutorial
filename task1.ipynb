```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data = pd.read_csv('mydata.csv')
data.describe()
```

```python
y = data['GPA']
x1 = data['SAT']
```

```python
plt.scatter(x1, y)
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.title('Scatterplot of GPA vs SAT', fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

```python
plt.scatter(x1, y)
yhat = 0.0017 * x1 + 0.275
plt.plot(x1, yhat, color='red', linewidth=2)
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.title('Scatterplot of GPA vs SAT', fontsize=20)
plt.show()
```
